```python
import pandas as pd

# Asegurar formato de fecha
df['date'] = pd.to_datetime(df['date'])

# --- PASO AVANZADO: INGENIERA DE CARACTERSTICAS CLIMTICAS ---
# Para probar "transici贸n", necesitamos el historial clim谩tico, no solo el dato del d铆a.
# Creamos un sub-dataset solo con el clima por Comuna y Fecha (sin duplicados de incendios)
df_clima = df[['date', 'comuna', 'temperature_2m_mean',
               'relativehumidity_2m_mean', 'windspeed_10m_mean',
               'precipitation_sum']].drop_duplicates().sort_values(['comuna', 'date'])

# Generamos variables de "d铆as anteriores" (Lags)
# Esto responde al criterio de "t茅cnicas de an谩lisis avanzadas"
cols_clima = ['temperature_2m_mean', 'relativehumidity_2m_mean', 'windspeed_10m_mean', 'precipitation_sum']

for col in cols_clima:
    # Clima de ayer (Lag 1) y anteayer (Lag 2)
    df_clima[f'{col}_lag1'] = df_clima.groupby('comuna')[col].shift(1)
    df_clima[f'{col}_lag2'] = df_clima.groupby('comuna')[col].shift(2)

    # Promedio m贸vil de 3 d铆as (captura tendencias acumuladas)
    df_clima[f'{col}_mean3d'] = df_clima.groupby('comuna')[col].rolling(window=3).mean().reset_index(level=0, drop=True)

# Unimos estas nuevas variables avanzadas al dataset original de incendios
df_final = df.merge(df_clima, on=['date', 'comuna'] + cols_clima, how='left')

# --- PREPARACIN DEL TARGET ---
# Filtramos solo filas donde hay incendio para analizar la intensidad
df_final = df_final[df_final['categoria_incendio'] != 'sin_incendio'].copy()

# Creamos la variable objetivo seg煤n tu hip贸tesis:
# 0: Baja Intensidad
# 1: Media o Alta Intensidad (Transici贸n ocurrida)
target_map = {'bajo': 0, 'medio': 1, 'alto': 1}
df_final['target_intensidad'] = df_final['categoria_incendio'].map(target_map)

# Limpieza final de nulos generados por los lags (los primeros d铆as no tienen "ayer")
df_final.dropna(inplace=True)

# Seleccionamos columnas relevantes para el modelo
columnas_modelo = [
    'target_intensidad',
    'latitud', 'longitud', 'frp', # Geogr谩ficas y F铆sicas
    'temperature_2m_mean', 'relativehumidity_2m_mean', 'windspeed_10m_mean', # Clima Hoy
    'temperature_2m_mean_lag1', 'temperature_2m_mean_mean3d', # Clima Ayer y Acumulado (NUEVAS)
    'relativehumidity_2m_mean_lag1', 'windspeed_10m_mean_lag1'
]

df_modelo = df_final[columnas_modelo]

# Vista previa y guardado
print(df_modelo.head())
print("Distribuci贸n del Target:", df_modelo['target_intensidad'].value_counts(normalize=True))

# Guardar
df_modelo.to_csv('dataset_final_incendios_biobio.csv',index=False)
```

```python
import pandas as pd

# Crear timestamp completo para ordenar eventos
df['acq_time_str'] = df['acq_time'].astype(str).str.zfill(4) # Formato HHMM
df['timestamp'] = pd.to_datetime(df['acq_date'] + ' ' + df['acq_time_str'], format='%Y-%m-%d %H%M', errors='coerce')

# 2. Agrupaci贸n Espacial Inteligente
# Redondeamos a 3 decimales (~110m) para agrupar detecciones del mismo foco que var铆an levemente en posici贸n
df['lat_group'] = df['latitude'].round(3)
df['lon_group'] = df['longitude'].round(3)

# Ordenar por ubicaci贸n y tiempo para detectar secuencias
df = df.sort_values(['lat_group', 'lon_group', 'timestamp'])

# 3. Identificaci贸n de "Eventos" nicos (L贸gica temporal)
# Si pasaron m谩s de 24 horas entre detecciones en el mismo sitio, es un nuevo incendio.
df['time_diff'] = df.groupby(['lat_group', 'lon_group'])['timestamp'].diff()
df['new_event'] = (df['time_diff'].dt.total_seconds() > 86400) | df['time_diff'].isna()
df['event_id'] = df['new_event'].cumsum() # ID 煤nico para cada racha de fuego

# 4. Construcci贸n del Dataset Final (Agregaci贸n)
def procesar_evento(grupo):
    # Datos del inicio del incendio (Condiciones iniciales)
    inicio = grupo.iloc[0]

    # C谩lculo de Intensidad
    frp_inicial = inicio['frp']
    frp_maximo = grupo['frp'].max()
    aumento = frp_maximo - frp_inicial

    # Definici贸n del Target seg煤n tu regla
    # Target 1: Hubo transici贸n (aumento > 3 MW)
    target = 1 if aumento > 3 else 0

    return pd.Series({
        'fecha_inicio': inicio['acq_date'],
        'latitud': inicio['latitude'],
        'longitud': inicio['longitude'],
        'comuna': inicio['comuna'],
        # Variables Clim谩ticas Iniciales (Las que "favorecen" la transici贸n)
        'temp_inicial': inicio['temperature_2m_mean'],
        'humedad_inicial': inicio['relativehumidity_2m_mean'],
        'viento_inicial': inicio['windspeed_10m_mean'],
        'precip_inicial': inicio['precipitation_sum'],
        # Variables del Fuego
        'frp_inicial': frp_inicial,
        'frp_maximo': frp_maximo,
        'delta_frp': aumento,
        'duracion_horas': (grupo['timestamp'].max() - grupo['timestamp'].min()).total_seconds() / 3600,
        'conteo_detecciones': len(grupo),
        # Target
        'target_transicion': target
    })

# Aplicar la agrupaci贸n
df_eventos = df.groupby('event_id').apply(procesar_evento).reset_index(drop=True)

# Filtrar ruidos (opcional: eventos con detecci贸n 煤nica no nos dicen mucho de transici贸n, pero son casos negativos v谩lidos)
# df_eventos = df_eventos[df_eventos['conteo_detecciones'] > 1]

# Guardar
print(df_eventos['target_transicion'].value_counts(normalize=True))
df_eventos.to_csv('Dataset_Incendios_Eventos_Target.csv',index=False)
```

```python
import pandas as pd

# Filtramos solo filas con datos de detecci贸n v谩lidos
df = df.dropna(subset=['acq_date', 'acq_time', 'latitude', 'longitude']).copy()

# 2. Crear Timestamp correcto
# Ajustamos el formato de hora HHMM para que sea legible
df['acq_time_clean'] = df['acq_time'].astype(int).astype(str).str.zfill(4)
df['timestamp'] = pd.to_datetime(df['acq_date'] + ' ' + df['acq_time_clean'], format='%Y-%m-%d %H%M')

# 3. Definir "Ubicaci贸n" (Binning Espacial)
# Redondeamos a 2 decimales (~1.1km) para agrupar pixeles cercanos del mismo incendio
df['lat_bin'] = df['latitude'].round(2)
df['lon_bin'] = df['longitude'].round(2)

# 4. Identificar Eventos nicos
df = df.sort_values(by=['lat_bin', 'lon_bin', 'timestamp'])
# Calculamos el tiempo entre detecciones en el mismo "pixel"
df['time_diff'] = df.groupby(['lat_bin', 'lon_bin'])['timestamp'].diff()
# Si es la primera detecci贸n o pasaron > 24h, es un nuevo evento
df['new_event'] = (df['time_diff'].isna()) | (df['time_diff'] > pd.Timedelta(days=1))
df['event_id'] = df['new_event'].cumsum()

# 5. Agrupar y Calcular M茅tricas (El paso clave)
df_agrupado = df.groupby('event_id').agg({
    'frp': ['first', 'max', 'count'],       # Intensidad inicial vs m谩xima
    'temperature_2m_mean': 'first',         # Clima al momento del inicio
    'relativehumidity_2m_mean': 'first',
    'windspeed_10m_mean': 'first',
    'lat_bin': 'first',
    'lon_bin': 'first',
    'acq_date': 'first'
}).reset_index()

# Aplanar nombres de columnas
df_agrupado.columns = ['event_id', 'frp_inicial', 'frp_max', 'n_detecciones',
                       'temp_inicial', 'humedad_inicial', 'viento_inicial',
                       'latitud', 'longitud', 'fecha']

# 6. Generar Target seg煤n tu regla
df_agrupado['delta_frp'] = df_agrupado['frp_max'] - df_agrupado['frp_inicial']
df_agrupado['target_transicion'] = (df_agrupado['delta_frp'] > 3).astype(int)

# Guardar
df_agrupado.to_csv('Dataset_Incendios_Agrupado_Final.csv',index=False)
```

```python
import pandas as pd
import numpy as np

# --- 1. CARGA DE DATOS ---
# Aseg煤rate de que el nombre del archivo coincida con el tuyo
archivo_entrada = '/content/sample_data/Clima_Incendios_Detallado_Clasificado.csv'
archivo_salida = 'Dataset_Incendios_Final_Completo_Lags.csv'

print(f"Cargando {archivo_entrada}...")
# Usamos low_memory=False para evitar advertencias de tipos mixtos
df = pd.read_csv(archivo_entrada, low_memory=False)

# Limpieza: eliminar filas sin datos de fecha/hora
df = df.dropna(subset=['acq_date', 'acq_time']).copy()

# Conversi贸n y limpieza de fechas
# Convertimos acq_time a string rellenando con ceros (ej. 130 -> '0130')
df['acq_time_clean'] = df['acq_time'].astype(float).astype(int).astype(str).str.zfill(4)
df['timestamp'] = pd.to_datetime(df['acq_date'] + ' ' + df['acq_time_clean'], format='%Y-%m-%d %H%M', errors='coerce')
df['date'] = pd.to_datetime(df['acq_date'])

# --- 2. GENERACIN DE VARIABLES CLIMTICAS CON LAGS (DAS ANTERIORES) ---
print("Generando variables de rezago (lags)...")

# Definimos las columnas de clima
cols_clima = ['temperature_2m_mean', 'relativehumidity_2m_mean', 'windspeed_10m_mean', 'precipitation_sum']

# Agrupamos por fecha y comuna para tener un registro 煤nico de clima diario por zona
# (Usamos mean() para colapsar duplicados si los hubiera, aunque suelen ser id茅nticos por d铆a/comuna)
df_clima = df.groupby(['date', 'comuna'])[cols_clima].mean().reset_index().sort_values(['comuna', 'date'])

# Creamos los lags de 1, 2 y 3 d铆as
for col in cols_clima:
    for i in range(1, 4): # 1, 2, 3
        df_clima[f'{col}_lag{i}'] = df_clima.groupby('comuna')[col].shift(i)

# --- 3. IDENTIFICACIN Y AGRUPACIN DE EVENTOS DE INCENDIO ---
print("Agrupando detecciones en eventos 煤nicos...")

# Filtramos solo filas con coordenadas y tiempo v谩lidos
df_fires = df.dropna(subset=['timestamp', 'latitude', 'longitude']).copy()

# Redondeo espacial (Binning) para agrupar pixeles cercanos (~1km)
df_fires['lat_bin'] = df_fires['latitude'].round(2)
df_fires['lon_bin'] = df_fires['longitude'].round(2)

# Ordenamos para detectar secuencias
df_fires = df_fires.sort_values(by=['lat_bin', 'lon_bin', 'timestamp'])

# Calculamos diferencia de tiempo entre detecciones en el mismo "pixel"
df_fires['time_diff'] = df_fires.groupby(['lat_bin', 'lon_bin'])['timestamp'].diff()

# Definici贸n de Nuevo Evento:
# Si es la primera detecci贸n (NaT) O si pasaron m谩s de 24 horas desde la anterior
df_fires['new_event'] = (df_fires['time_diff'].isna()) | (df_fires['time_diff'] > pd.Timedelta(days=1))
df_fires['event_id'] = df_fires['new_event'].cumsum()

# --- 4. CLCULO DE MTRICAS POR EVENTO (AGREGACIN) ---
df_agrupado = df_fires.groupby('event_id').agg({
    'frp': ['first', 'max'],          # Intensidad inicial vs m谩xima
    'timestamp': ['min', 'max'],      # Para duraci贸n
    'comuna': 'first',                # Comuna (solicitado)
    'lat_bin': 'first',
    'lon_bin': 'first',
    'date': 'first'                   # Fecha de inicio para cruzar con clima
}).reset_index()

# Aplanar nombres de columnas (quitar multi-铆ndice)
df_agrupado.columns = ['event_id', 'frp_inicial', 'frp_max', 'time_start', 'time_end',
                       'comuna', 'latitud', 'longitud', 'fecha_inicio']

# Calcular variables derivadas
df_agrupado['duracion_horas'] = (df_agrupado['time_end'] - df_agrupado['time_start']).dt.total_seconds() / 3600
df_agrupado['delta_frp'] = df_agrupado['frp_max'] - df_agrupado['frp_inicial']

# DEFINICIN DEL TARGET (HIPTESIS)
# 1 si el fuego creci贸 m谩s de 3 MW, 0 si no.
df_agrupado['target_transicion'] = (df_agrupado['delta_frp'] > 3).astype(int)

# --- 5. UNIN FINAL (MERGE) ---
print("Uniendo clima hist贸rico al dataset de eventos...")
# Unimos los datos del incendio con el clima hist贸rico (lags) usando Comuna y Fecha
df_final = df_agrupado.merge(df_clima, left_on=['comuna', 'fecha_inicio'], right_on=['comuna', 'date'], how='left')

# Limpieza final
df_final = df_final.drop(columns=['date']) # Redundante

# --- 6. GUARDADO ---
df_final.to_csv(archivo_salida, index=False)
print(f"隆xito! Archivo guardado como: {archivo_salida}")
print(f"Total de eventos encontrados: {len(df_final)}")
print(f"Distribuci贸n del Target: \n{df_final['target_transicion'].value_counts()}")
```
